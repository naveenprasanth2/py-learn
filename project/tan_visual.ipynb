```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range, avoiding the points where tan(x) is undefined
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y_sin = np.sin(x)
y_cos = np.cos(x)

# For tan(x), mask values where cos(x) is close to zero to avoid plotting vertical asymptotes
y_tan = np.tan(x)
y_tan[np.abs(y_cos) < 0.1] = np.nan  # Avoid plotting where tan(x) "blows up"

plt.figure(figsize=(10, 6))
plt.plot(x, y_sin, label='sin(x)', linewidth=2)
plt.plot(x, y_cos, label='cos(x)', linewidth=2)
plt.plot(x, y_tan, label='tan(x)', linestyle='--', color='red', linewidth=1.5)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-3, 3)
plt.xlim(x[0], x[-1])

plt.title('sin(x), cos(x), and tan(x)')
plt.xlabel('x (radians)')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
